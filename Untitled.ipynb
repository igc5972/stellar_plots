```python
import PIL
from PIL import Image
from astropy.io import fits
from matplotlib import pyplot as plt
from matplotlib import cm


#Open the image as an array
fits_img = fits.open('/path/to/img.fits')
orig_img = fits_img[0].data


#Normalize the array
orig_img /= np.max(orig_img)


#Convert array to PIL format
orig_im_pil = Image.fromarray(np.uint8(cm.viridis(orig_img)*255))


#Resize image (and set a resampling method that is superior!)
resized = orig_im_pil.resize((100,100), Image.Resampling.LANCZOS)


#Compare the two outputs
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(orig_im_pil)
ax1.set_title('Original')
ax2.imshow(resized)
ax2.set_title('Resized')
plt.show()



#Get the resized image back as an array
resized_arr = np.asarray(resized)


#Check the shape -- oops, there are four channels (dimensions) (RGBA), let's fix that
print(resized_arr.shape)


#
resized_img_gray = resized.convert('L')


#Now let's convert to array and check shape -- all good!
fixed_resized_arr = np.asarray(resized_img_gray)

print(fixed_resized_arr.shape)



#Let down, the pixel values are not going to have scaled the right way
#They are going to be normalized 0 - 255
```
